# <span class="tema">(Cerca)</span> Aplicació dels algorismes genètics a la cerca del màxim d'una funció multimodal

```python
# Creem la població inicial
def initpop(n,long):
    import random
           # Generem una poblacio de n cromosomes de longitud long. 
    pop = [[0] * long for x in range(n)]
    for i in range(n):
        for j in range(long):
            if random.random()>0.5: pop[i][j] += 1 
    return pop
```

```python
poblacio1 = initpop(50,22)
print(poblacio1)
```

```python
# Definim la funció d'adaptació 
# d’un cromosoma de len(r) bits.
def fit(r):
    import math
           # Transformem els bits en un valor real 
           # a l'interval [-1,2]
    sum=0.0
    for i in range(len(r)):
        sum = sum + r[i]*(2**i)
    x = -1.0 + sum * (3.0/(2.0**(len(r))-1.0))
           # Avaluem el cromosoma
    y = x * math.sin(10*math.pi*(x))+1.0
    return y
```

```python
fit(poblacio1[0])
```

```python
for i in range(50):
    print(fit(poblacio1[i]))
```

```python
# Definim la mutació amb probabilitat mutprob
def mutacio(r,mutprob):
    import random
    for i in range(len(r)):
        if random.random() < mutprob: 
            if r[i]==0:
                r[i]=1
            else:
                r[i]=0
    return r
```

```python
pm = 0.01
mutacio(poblacio1[0],pm)
```

```python
for i in range(50):
    print(poblacio1[i])
    print(mutacio(poblacio1[i],pm))
    print()
```

```python
def creuament(r1,r2):
    import random
    i=random.randint(1,len(r1)-2)
    return r1[:i]+r2[i:],r1[i:]+r2[:i]
```

```python
print(poblacio1[0])
print(poblacio1[1])
creuament(poblacio1[0],poblacio1[1])
```
